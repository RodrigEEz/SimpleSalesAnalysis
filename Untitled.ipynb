### Task 1: concatenate every csv into one DataFrame

```python
# import all necessary libs and files
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
from os import listdir
from os.path import isfile, join

plt.style.use('seaborn')

mypath = '/home/sebastian/Projects/SalesAnalysis/Sales_Data'

data = [f for f in listdir(mypath) if isfile(join(mypath, f))]
```

```python
#concatenate every dataset into one
yearly_data = pd.DataFrame()

for info in data:
    yearly_data = pd.concat([yearly_data, pd.read_csv(mypath + '/' + info)], axis=0)
    
    
#change column names
yearly_data.rename(columns={'Order Date':'Date', 
                                          'Quantity Ordered' : 'Quantity',
                                         'Price Each': 'Price',
                                         'Purchase Address' : 'Adress',
                           'Order ID': 'ID'}, inplace=True)
    
yearly_data.head()
```

#### Clean  and transform the data

```python
#drop missing values
yearly_data.dropna(inplace=True)

#drop ambiguous rows
yearly_data = yearly_data[yearly_data['Date'].str[0:2] != 'Or']

#transform data
yearly_data['Price'] = yearly_data['Price'].astype('float64')
yearly_data['Quantity'] = yearly_data['Quantity'].astype('float64')
```

### Task 2: Find the best and month for sales and show how much was earned 

```python
#Change date column type to datetime
yearly_data['Date'] = pd.to_datetime(yearly_data['Date'])

#Make year and month column
yearly_data['Month'] = yearly_data['Date'].dt.month

yearly_data.head()

#Add total sale value column
yearly_data['Total_Sale'] = yearly_data['Quantity'] * yearly_data['Price']

yearly_data.head()
```

```python
#group by month & by year
sales_by_month = yearly_data.groupby('Month')['Total_Sale'].sum()
sales_by_month.head()

# print which was the month with most revenue
max_revenue = sales_by_month.max()

print('the month with most revenue as of 2019 sales was December with ${:,.2f} of total revenue'.format(max_revenue))
```

```python
#plot sales by month
fig, ax = plt.subplots()
ax.bar(range(1,13),sales_by_month)
ax.get_yaxis().set_major_formatter(matplotlib.ticker.FuncFormatter(lambda x,p: format(int(x), ',')))
ax.set_xlabel('Month')
ax.set_ylabel('revenue')
plt.title('sales by month')
plt.show()
```

### Task 3: Show the city with most sales and how much was the revenue


```python
#make a city column
yearly_data['City'] = yearly_data['Adress'].apply(lambda x: x.split(',')[1])

yearly_data.head()
```

```python
# get sales by city
sales_by_city = yearly_data.groupby('City')['Total_Sale'].sum()

# print the city with most sales and its total revenue
max_revenue = sales_by_city.max()
city_max_revenue = sales_by_city.idxmax()

print('The city with most sales was ' + city_max_revenue + ' with ${:,.2f} of total revenue'.format(max_revenue))
```

```python
#  plot sales by city
fig, ax = plt.subplots()
ax.bar(sales_by_city.index, sales_by_city)
plt.xticks(rotation='vertical')
ax.get_yaxis().set_major_formatter(matplotlib.ticker.FuncFormatter(lambda x,p: format(int(x), ',')))
ax.set_xlabel('City')
ax.set_ylabel('Price (In US Dollars)')
plt.title('Sales by city')
```

### Task 4 : Show the most sold and most expensive item 

```python
#group by Product
items_sold = yearly_data.groupby('Product')['Quantity'].sum()
items_revenue = yearly_data.groupby('Product')['Price'].mean()
```

```python
fig, ax1 = plt.subplots()

ax1.bar(items_sold.index, items_sold, color='g')
ax1.set_ylabel('quantity', color='g')

ax2 = ax1.twinx()
ax2.plot(items_revenue.index, items_revenue)
ax2.set_ylabel('price (in US dollars)', color='b')
ax2.set_xlabel('Product')
ax2.grid(None)

products = yearly_data.groupby('Product')

items = []

for item, value in products:
    items.append(item)

ax1.set_xticklabels(items, rotation='vertical', size=8)
plt.title('price and quantity sold by product')
```

### Task 5: Show which are the best hours for advertisement


```python
# create an hour column
yearly_data['Hour'] = yearly_data['Date'].dt.hour
yearly_data.head()
```

```python
# create a sales per hour df
sales_per_hour = yearly_data.groupby('Hour')['Total_Sale'].sum()

fig, ax = plt.subplots()
ax.plot(sales_per_hour.index, sales_per_hour)
ax.set_ylabel('sales (in US dollars)')
ax.set_xlabel('hour')
ax.get_yaxis().set_major_formatter(matplotlib.ticker.FuncFormatter(lambda x,p: format(int(x), ',')))
plt.title('Sales by hour')
```
